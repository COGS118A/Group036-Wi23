```python
import pandas as pd 
import numpy as np 
from joblib import dump, load
```

```python
#read_csv into file
full_features = pd.read_csv('full_features.csv')
#drop yield_class from dataset
full_features = full_features.drop(['yield_class'], axis = 1)
```

```python
#SVM model important features

# Load the saved model
model = load('svm_model.pkl')

# Get feature importance scores and corresponding feature names
feature_importance = abs(model.named_steps['clf'].coef_[0])
feature_names = full_features.columns

# Create a list of tuples containing feature names and importance scores
feature_scores = list(zip(feature_names, feature_importance))

# Sort the list by importance score in descending order
feature_scores.sort(key=lambda x: x[1], reverse=True)

# Print feature names and importance scores
print("\nFeature Importance Scores:\n")
for feature, score in feature_scores:
    print(feature,':\n', score)
```
